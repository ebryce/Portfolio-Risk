```python
rebal_px = pd.read_sql('''
    SELECT s.date AS rebal_date, s.ticker, c AS rebal_px,
    CASE WHEN flag=0 THEN -1 ELSE 1 END AS flag
    FROM cfmrc s
    INNER JOIN indexchanges i
    WHERE s.date=i.date 
        AND s.ticker=i.ticker'''
    ,conn)

all_px = pd.read_sql('''
    SELECT date,ticker, MAX(c) AS px
    FROM cfmrc
    GROUP BY ticker,date'''
    ,conn).drop_duplicates()
```

```python
merged = rebal_px.merge(all_px, left_on=['ticker'], right_on=['ticker'], how='right').sort_values(by=['ticker','date'])
merged['flag'].fillna(0, inplace=True)
merged['flag'] = (merged['date']==merged['rebal_date'])*merged['flag']
merged['rebal_date'] = merged['rebal_date'].apply(lambda date: datetime.strptime(date, '%Y-%m-%d') if type(date)==str else np.nan)
```

```python
import requests

def tgam_html(response):
    html = str(response.content)[2:]
    table = []

    dates = html.split('<tr>')[1].split('<th')[1:]
    dates = [date[date.find('>')+1:] for date in dates]
    dates = [date[:date.find('<')] for date in dates]
    dates = dates[1:]
    dates = [datetime.strptime(date,'%m-%Y') for date in dates]

    rows = html.split('<tr')[2:]
    rows_temp = []
    n = None

    statement = []
    header = None
    
    for row in rows:

        cells = row.split('<td')

        cells = [cell[cell.find('>')+1:] for cell in cells]
        cells = [cell[:cell.find('<')] for cell in cells]

        if cells[0]=='':
            cells = cells[1:]

        statement_i = pd.Series(name=cells[0].replace('&#039;','').replace('&amp;','and'))
        cells = cells[1:]

        # Fix numbers
        cells = [cell.replace('M','000') for cell in cells]
        cells = [cell.replace(' ','') for cell in cells]
        cells = [cell.replace(',','') for cell in cells]
        cells = [cell.replace('&nbsp;','N/A') for cell in cells]
        cells = [np.nan if cell=='N/A' else float(cell) for cell in cells]

        for i in range(len(cells)):
            statement_i[dates[i]]=cells[i]
            
        if (statement_i.name=='&nbsp;'):
            pass
        elif ((pd.DataFrame(statement_i).isna().sum()/pd.DataFrame(statement_i).isna().count()<1).iloc[0]):
            if statement_i.name=='Total':
                statement_i.name = ' '.join([statement_i.name, header])
            statement.append(pd.DataFrame(statement_i).T)
        else:
            header = statement_i.name
            pass
        
    statement = pd.concat(statement)
    return statement

def tgam_financials(ticker, quarterly=False):
    ticker = ticker.replace('.','-')+'-T'
    
    table = []
    for statement in ['incomeStatement','cashFlow','balanceSheet']:
        url = 'https://globeandmail.pl.barchart.com/module/financials/{Statement}.html'.format(Statement=statement)
        
        period = '3m' if quarterly else '12m'
        
        r = requests.post(url, data={"period":period,"symbol":ticker,"showPeriodTabs":0})
        table.append(tgam_html(r))
        
    table = pd.concat(table).T.reset_index()
    table['period'] = 'Q' if quarterly else 'A'
    table['ticker'] = ticker[:ticker.find('-')]
    return table
```

```python
tickers = list(merged.loc[merged['flag']!=0]['ticker'].unique())
quarterlies = [True, False]
tables = []
failed = []

reload = 'tgam.h5' not in os.listdir()
if reload:
    for ticker in tickers:
        try:
            for quarterly in quarterlies:
                    t = tgam_financials(ticker=ticker, quarterly=quarterly)
                    t['ticker']=ticker
                    t['quarterly']=quarterly
                    tables.append(t)
        except:
            print(ticker,'failed')
            if '.' in ticker:
                retry_ticker = ticker[:ticker.find('.')]
                print('> trying', retry_ticker)
                tickers.append(retry_ticker)
            failed.append(ticker)
            continue

    all_cols = []
    for t in tables:
        all_cols.extend(t.columns)
    all_cols = list(set(all_cols))

    for t in tables:
        for c in all_cols:
            if c not in t.columns:
                t[c] = np.nan
        t.drop(t.loc[:,t.columns.duplicated()],axis=1, inplace=True)
    tables = pd.concat(tables, sort=False)
    tables.to_hdf('tgam.h5', key='tgam')
else:
    tables = pd.read_hdf('tgam.h5', key='tgam')
```

```python
columns = list(tables.columns)
for key in ['index','period','ticker']:
    columns.remove(key)
for c_i in columns:
    for c_j in columns:
        if c_i!=c_j:
            try:
                tables['-'.join([c_i, c_j])]= tables[c_i]/tables[c_j]
            except:
                print('Cannot feature-engineer %s' % ','.join([c_i,c_j]))
                
full['start'] = (full['index']-pd.Timedelta(28, 'D')).apply(lambda date: date.replace(day=1))
full = full.merge(merged.loc[merged['flag']!=0][['ticker','rebal_date','flag']].drop_duplicates(),on=['ticker'], how='left')

full['rebal_period'] = ((full['rebal_date']>full['start']) & (full['rebal_date']<full['index']))
full['Q'] = ((full['index']-full['rebal_date'])/pd.Timedelta(365/4, 'D')).apply(lambda Q: int(math.ceil(Q)) if Q==float(Q) else np.nan)
full['A'] = ((full['index']-full['rebal_date'])/pd.Timedelta(365, 'D')).apply(lambda A: int(math.ceil(A)) if A==float(A) else np.nan)
```

```python
df = full.copy()
df['rebal_next_period'] = ((df['Q']==0) & (df['period']=='Q')) | ((df['A']==0) & (df['period']=='A'))
df['flag_next_period'] = df['rebal_next_period']*df['flag']
df['drop_next_period'] = df['flag_next_period']==-1
df['add_next_period'] = df['flag_next_period']==1

df = df.merge(tables, left_on=['index','quarterly','ticker'], right_on=['index','quarterly','ticker'], how='outer', suffixes=('','_t'))
```

## Optimize sample

```python
from scipy.optimize import curve_fit

def sigmoid(x, x0, k):
    y = 1 / (1 + np.exp(-k*(x-x0)))
    return y

nulls = pd.DataFrame(tables.count()/tables.count().max()).sort_values(by=0,ascending=False)

Y = nulls[0]
X = np.linspace(0,1, len(Y))
popt, pcov = curve_fit(sigmoid, X,Y)
Y_est = sigmoid(X, *popt)

plt.scatter(X, Y, color='red', s=3)
plt.plot(X, Y_est, color='blue')
plt.xlabel('Cum. % Rows')
plt.ylabel('% with Factor')
plt.ylim(0,1)
plt.xlim(0,1)
plt.show()

#####

threshold = np.linspace(0,1,100)
responses = []
for thresh in threshold:
    # Drop columns with too much NAN
    full = tables.dropna(thresh=thresh*len(tables.index), axis=1)
    # Drop rows with too much NAN
    full = full.dropna(axis=0)
    responses.append(len(full.index)/len(tables.index))

plt.scatter(threshold, responses, color='red', s=3)
plt.xlabel('Number of columns to drop')
plt.ylabel('Rows remaining after cut')
plt.ylim(0,1)
plt.xlim(0,1)
plt.show()

def thresh(df, pct, axis):
    if axis==0:
        return int(round(len(df.index)*pct))
    elif axis==1:
        return int(round(len(df.columns)*pct))
    else:
        return 0

xrows = []
xcols = []
n = 10
for x in np.linspace(0,0.2,n):
    print(x,'of',n)
    xcol = pd.Series(name=x)
    xrow = pd.Series(name=x)
    for y in np.linspace(0,0.2,n):
        #print(' ',y)
        sani = df.dropna(axis=0, thresh=thresh(df=df, pct=x, axis=0)).dropna(axis=1, thresh=thresh(df=df, pct=y, axis=1)).dropna(axis=0)
        xrow[y]=len(sani.index)
        xcol[y]=len(sani.columns)
    xcols.append(pd.DataFrame(xcol).T)
    xrows.append(pd.DataFrame(xrow).T)
xcols = pd.concat(xcols)
xrows = pd.concat(xrows)

test = pd.DataFrame(xcols.iloc[0,:]).reset_index().merge(pd.DataFrame(xrows.iloc[0,:]).reset_index(), on='index').drop(['index'],axis=1)
plt.plot(test[test.columns[0]],test[test.columns[1]])
plt.xlabel('Number of columns')
plt.ylabel('Number of rows')
plt.show()
```

```python
def get_mlp():
    nn_model = Sequential()
    nn_model.add(Dense(units=256, activation='relu', input_dim=len(X.columns), kernel_initializer='normal'))
    nn_model.add(Dense(units=256, activation='relu', kernel_initializer='normal'))
    nn_model.add(Dense(units=256, activation='relu', kernel_initializer='normal'))
    nn_model.add(Dense(units=256, activation='relu', kernel_initializer='normal'))
    nn_model.add(Dense(units=256, activation='relu', kernel_initializer='normal'))
    nn_model.add(Dense(units=256, activation='relu', kernel_initializer='normal'))
    nn_model.add(Dense(units=256, activation='relu', kernel_initializer='normal'))
    nn_model.add(Dense(units=256, activation='relu', kernel_initializer='normal'))
    nn_model.add(Dense(units=256, activation='relu', kernel_initializer='normal'))
    nn_model.add(Dense(units=1, kernel_initializer='normal'))
    nn_model.compile(loss='mean_squared_error', optimizer='adam')
    return nn_model

nn_model = get_mlp()
fit = nn_model.fit(X_train.values, y_train.values, epochs=100, batch_size=128, verbose=1)

plt.plot(nn_model.history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.title('Multi-layer Perceptron Loss')

print('Minimum Mean Squared Error: {0}'.format(min(nn_model.history.history['loss'])))

y_pred = nn_model.predict(X_test)
print('Explained Variance Score: {0}'.format(explained_variance_score(y_pred, y_test)))
print('Mean Squared Error: {0}'.format(mean_squared_error(y_pred, y_test)))

#clf_model_eval(y_test, y_pred)
```
