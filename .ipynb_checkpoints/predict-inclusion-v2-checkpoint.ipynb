## Load data from the database

```python
date_range = list(pd.read_sql('''SELECT MIN(date), MAX(date) FROM fact''', conn).values[0])

df_changes = pd.read_sql('''
    SELECT *
    FROM factset_index_changes
    WHERE date BETWEEN (SELECT MIN(date) FROM fact) AND (SELECT MAX(date) FROM fact)''',
    conn)

df_changes['Y'] = df_changes['date'].apply(lambda date: int(date[:4]))
df_changes['Q'] = df_changes['date'].apply(lambda date: math.ceil(int(date[5:7])/3))
df_changes['flag'] = df_changes['action'].apply(lambda action: 1 if action=='Add' else -1)

df_ratios = pd.read_sql('''
    SELECT *
    FROM ratios''',
    conn)

indices = df_changes['index'].unique()
```

### Find the date range

```python
def get_last_n_quarters(Y, Q, n, fmt=str):
    y,q = Y,Q
    last_n_quarters = []
    for i in range(n):
        q -= 1
        if q == 0:
            q = 4
            y -= 1
        if fmt==int:
            last_n_quarters.append([q,y])
        else:
            last_n_quarters.append('{Y}_Q{Q}'.format(Y=y, Q=q))
    return last_n_quarters

quarter_range = []
for column in df_ratios.columns:
    date = column[column.find('_')+1:]
    if len(date)!=7:
        continue
    if len(quarter_range)<2:
        quarter_range.append(date)
    elif date < min(quarter_range):
        quarter_range[quarter_range.index(min(quarter_range))]=date
    elif date > min(quarter_range):
        quarter_range[quarter_range.index(max(quarter_range))]=date

d_year = int(max(quarter_range)[:4])-int(min(quarter_range)[:4])
max_n_quarters = int(max(quarter_range)[6:])-int(min(quarter_range)[6:]) + d_year*4 + 1
last_n_quarters = get_last_n_quarters(Y=int(max(quarter_range)[:4]), Q=int(max(quarter_range)[6:]), n=max_n_quarters, fmt=str)
print('There are ratios for a maximum of %d quarters, from %s to %s:\n%s' % (max_n_quarters, min(quarter_range), max(quarter_range), ', '.join(last_n_quarters)))
```

### Index accounting ratio dates to the rebalance change

```python
ratios = []
for column in df_ratios.columns:
    if '_' in column:
        ratio = column[:column.find('_')]
        if ratio not in ratios and ratio != 'Ticker':
            ratios.append(ratio)
print('Ratios: %s'%', '.join(ratios))
rows = []
no_accounting_data_for = []
no_change_data_for = []
for ticker in df_changes.loc[df_changes['index']==index_of_interest].ticker.unique():
    # For each ticker, find the appropriate accounting ratios
    ratio_series_both = df_changes.loc[(df_changes['index']==index_of_interest) & (df_changes['ticker']==ticker)]
    for action in ratio_series_both['action'].unique():
        # Seperate by directionality, in case the name was added and then removed afterwards
        change = ratio_series_both.loc[ratio_series_both['action']==action]
        if len(change.index)>0:
            # Assuming there is accounting data for this name
            change = change.iloc[0]
            ratio_series = df_ratios.loc[(df_ratios['Ticker']==ticker) & (df_ratios['Exchange']==change['exchange'])]
            # Walk everything back to the current quarter
            if len(ratio_series.index)>0:
                ratio_columns = []
                for ratio in ratios:
                    for nth_quarter in last_n_quarters:
                        column_name = ratio+'_'+nth_quarter
                        if column_name in ratio_series.columns:
                            ratio_columns.append(column_name)
                this_ratios = ratio_series[ratio_columns]
                relative_column_names = []
                for column in this_ratios.columns:
                    for nth_quarter in last_n_quarters:
                        if nth_quarter in column:
                            column = column.replace(nth_quarter, str(last_n_quarters.index(nth_quarter)))
                    relative_column_names.append(column)
                this_ratios.columns = relative_column_names
                
                row = pd.concat([ pd.DataFrame(change).T, this_ratios], axis=0)
                row.iloc[1] = row.iloc[1].fillna(row.iloc[0])
                rows.append(pd.DataFrame(row.iloc[1]).T)
                
            else:
                no_accounting_data_for.append([action, ticker])
                pass
        else:
            no_change_data_for.append([action, ticker])
            pass
        
rows = pd.concat(rows)

if len(no_change_data_for)>0:
    print('No change data for: %s' % ', '.join(['%s on %s'%(d[0], d[1]) for d in no_change_data_for]))
if len(no_accounting_data_for)>0:
    print('No accounting data for: %s' % ', '.join(['%s on %s'%(d[0], d[1]) for d in no_accounting_data_for]))

display(rows.head())
```
